```python
import pyautogui
import time
from datetime import datetime
#alerta
pyautogui.alert('O código irá rodar a partir de agora, NÃO PRESSIONE NENHUMA TECLA OU MEXA O MOUSE ATÉ O TERMINO')
pyautogui.PAUSE = .5

def EntraMateria(mouseX, mouseY):
    pyautogui.moveTo(mouseX, mouseY)
    pyautogui.mouseDown()
    pyautogui.mouseUp()
def ChecaHora():
    horaString = datetime.now().strftime('%H:%M')
    hora, minuto = horaString.split(':')
    horaFloat = int(hora) + (int(minuto) / 60)
    return horaFloat
    
#checando dia da semana
diaSemana = datetime.today().weekday()
if diaSemana != 5 and diaSemana!=6:
   
    #abrir navegador
    pyautogui.press('winleft')
    pyautogui.write('chrome')
    pyautogui.press('enter')

    #abrir gmail
    pyautogui.write('https://gmail.com/')
    pyautogui.press('enter')
    
    #abrir nova guia
    pyautogui.hotkey('ctrl', 't')

    #abrir AVA
    pyautogui.write('https://ava.unicarioca.edu.br/')
    pyautogui.press('enter')
    pyautogui.moveTo(828, 662)
    time.sleep(1)
    pyautogui.mouseDown()
    pyautogui.mouseUp()
    pyautogui.moveTo(913, 745)
    pyautogui.mouseDown()
    pyautogui.mouseUp()
    time.sleep(5)
    #abrir matéria do dia no AVA
    horaAtual = ChecaHora()
    if diaSemana == 0:
        if horaAtual < 19.40: 
            EntraMateria(665, 940)
        else:
            pyautogui.scroll(-500)
            EntraMateria(625, 929)
            
    elif diaSemana == 1:
        if horaAtual < 20.50:
            pyautogui.scroll(-500)
            EntraMateria(664, 735)
        else:
            pyautogui.scroll(-1200)
            EntraMateria(642, 947)
    elif diaSemana == 2:
        #sem checagem de hora, pois tem uma matéria no dia
        pyautogui.scroll(-1200)
        EntraMateria(639, 736)
        
    elif diaSemana == 4:
        #sem checagem de hora, pois tem uma matéria no dia
        pyautogui.scroll(-1200)
        EntraMateria(661, 541)

    #abrir nova guia
    pyautogui.hotkey('ctrl', 't')

    #abrir Linkedin
    pyautogui.write('https://www.linkedin.com/feed/')
    pyautogui.press('enter')

    #abrir nova guia
    pyautogui.hotkey('ctrl', 't')

    #abrir soundcloud
    pyautogui.write('https://soundcloud.com/you/history')
    pyautogui.press('enter')



elif diaSemana == 5 or diaSemana == 6:
   #abrindo paint tool sai 
   pyautogui.press('winleft')  
   pyautogui.write('paint tool sai')
   pyautogui.press('enter')
   pyautogui.hotkey('ctrl', 'o')
   pyautogui.moveTo(459, 909)
   pyautogui.mouseDown()
   pyautogui.mouseUp()
   pyautogui.write('billie eilish 2021 remake.sai')
   pyautogui.press('enter') 

pyautogui.alert('O código acabou de ser executado :)')
```
